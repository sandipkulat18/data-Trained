# Spliting the dataframe

```python
lr=LinearRegression()
for i in  range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print('At Random state',i,'the model perform Well')
        print('At Random State',i)
        print('Training accuncy score is ',round(accuracy_score(y_train,pred_train)*100,1))
        print('Testing accuncy score is ',round(accuracy_score(y_test,pred_test)*100,1),'\n\n')
```

```python
def model_selection(algorithum,x_train,y_train,x_test,y_test):
    algorithum.fit(x_train,y_train)
    model_1_pred_train=algorithum.predict(x_train)
    model_1_pred_test=algorithum.predict(x_test)
    print(accuracy_score(y_train,model_1_pred_train))
    print(accuracy_score(y_test,model_1_pred_test))
    print(classification_report(y_test,model_1_pred_test))
    print(classification_report(y_train,model_1_pred_train))
    print('Confusion matrics:\n', confusion_matrix(y_test,model_1_pred_test))
    
```
