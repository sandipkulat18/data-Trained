```python
from sklearn.preprocessing import MinMaxScaler
mns=MinMaxScaler()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
features=df.drop('pH',axis=1)
target=df['pH']
```

```python
for i in range(0,100):
    features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=i)
    lr.fit(features_train,target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)
    print(f'At random stat {i}, the training accurancy is :-{r2_score(target_train,pred_train)} ')
    print(f'At random stat {i}, the training accurancy is :-{r2_score(target_test,pred_test)} ')
    print('\n')
          
```

```python
 features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=12)
```

```python
lr.fit(features_train,target_train)
```

```python
pred_test=lr.predict(features_test)
```

```python
print(r2_score(target_test,pred_test))
```

Cross Validation Method

```python
Train_accurancy=r2_score(target_train,pred_train)
Test_accurancy=r2_score(target_test,pred_test)
```

```python
from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cv_score=cross_val_score(lr,features,target,cv=j)
    cv_mean=cv_score.mean()
    print(f'At cross fold {j},the cv score is {cv_mean} and accurancy score  for training is {Train_accurancy} and accurancy score for testing is {Test_accurancy}')
    print('\n')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,7))
plt.scatter(x=target_test,y=pred_test,color='r')
plt.plot(target_test,target_test,color='b')
plt.xlabel('Actual Label',fontsize=10)
plt.ylabel('prediction Charges', fontsize=0)
plt.title('Linear Regression', fontsize=10)
plt.savefig('lr.png')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters={'criterion':['mse','mae'],
           'max_features':['auto','sqrt','log2']}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(features_train,target_train)
print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mse',max_features='auto')
rf.fit(features_train,target_train)
rf.score(features_train,target_train)
pred_decision=rf.predict(features_test)

rfs=r2_score(target_test,pred_decision)
print('R2 Score :',rfs*100)
rfscore=cross_val_score(rf,features,target,cv=5)
rfc=rfscore.mean()
print('cross Val score :',rfc*100)
```
